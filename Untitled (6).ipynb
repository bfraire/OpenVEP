```python
import numpy as np
import os
from scipy.signal import butter, filtfilt

# Sampling Parameters
MAX_EEG_CHANNELS = 8
MAX_SAMPLES = 32706  # Set a standard length
FS = 250  # Sampling Frequency (Hz)

# Recursive file search function
def find_npy_files(directory):
    npy_files = []
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith(".npy"):
                npy_files.append(os.path.join(root, file))
    return npy_files

# Function to standardize EEG data shape
def standardize_shape(data, max_channels=MAX_EEG_CHANNELS, max_samples=MAX_SAMPLES):
    num_channels, num_samples = data.shape
    
    # Adjust channels
    if num_channels < max_channels:
        pad_width = ((0, max_channels - num_channels), (0, 0))
        data = np.pad(data, pad_width, mode='constant')
    elif num_channels > max_channels:
        data = data[:max_channels, :]
    
    # Adjust samples
    if num_samples < max_samples:
        pad_width = ((0, 0), (0, max_samples - num_samples))
        data = np.pad(data, pad_width, mode='constant')
    elif num_samples > max_samples:
        data = data[:, :max_samples]

    return data

# Bandpass filter function (0.5 - 50 Hz)
def bandpass_filter(data, lowcut=0.5, highcut=50.0, fs=FS, order=5):
    nyq = 0.5 * fs  # Nyquist frequency
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, data)  # Apply forward and backward filter

# Load and preprocess EEG data
def load_and_preprocess(base_directory):
    data_list = []
    labels_list = []
    groups_list = []
    file_count = 0

    for trial_num in range(1, 7):  # Assuming 6 trials
        trial_dir = os.path.join(base_directory, f"trial{trial_num}")
        print(f"Checking: {trial_dir}")

        # Search for .npy files recursively
        npy_files = find_npy_files(trial_dir)

        if not npy_files:
            print(f" No .npy files found in {trial_dir}")
            continue
        
        for file_path in npy_files:
            raw_data = np.load(file_path)

            if raw_data.ndim != 2:
                print(f" Skipping {file_path}: Expected 2D array but got {raw_data.shape}")
                continue

            print(f" Loaded {file_path} with shape {raw_data.shape}")

            # Standardize data shape
            processed_data = standardize_shape(raw_data)

            # Apply bandpass filtering along each channel
            filtered_data = np.apply_along_axis(bandpass_filter, axis=1, arr=processed_data)

            data_list.append(filtered_data)

            # Assign alternating labels (ensuring two classes)
            label = 0 if file_count % 2 == 0 else 1  
            labels_list.append(label)

            # Assign the trial number as the group ID
            groups_list.append(trial_num)  
            file_count += 1

    if file_count == 0:
        print(" No valid EEG files were processed!")
        return None, None, None

    # Convert to NumPy arrays
    X = np.array(data_list, dtype=np.float64)
    y = np.array(labels_list)
    groups = np.array(groups_list)

    print(f" Successfully processed {file_count} EEG files.")
    print(f" Final data shape: {X.shape}, Labels shape: {y.shape}, Groups shape: {groups.shape}")
    return X, y, groups

# Run the function to load, standardize, and filter the EEG data
directory_path = 'breanna-data'
X, y, groups = load_and_preprocess(directory_path)

# Verify results
if X is not None:
    print(f" Data Loaded: {X.shape}, Labels: {y.shape}, Groups: {groups.shape}")
```

```python
import numpy as np

# Define the correct path
eeg_file_path = "breanna-data/trial1/cyton8_alternating-vep_32-class_1.2s/sub-01/ses-01/eeg_2-per-class_run-1.npy"

# Load the EEG data
eeg_data = np.load(eeg_file_path)

# Confirm shape
print(f"EEG Data Shape: {eeg_data.shape}")

# Select a valid channel for FFT (e.g., channel 2)
channel_data = eeg_data[1]  # Assuming EEG data is structured as (channels, samples)

# Now rerun your FFT analysis
from scipy.fft import fft, fftfreq
import matplotlib.pyplot as plt

# FFT parameters
N = len(channel_data)  # Number of samples
T = 1.0 / 250.0        # Sampling interval (250 Hz)

# Compute FFT
fft_values = fft(channel_data)
frequencies = fftfreq(N, T)

# Plot frequency spectrum
plt.figure(figsize=(10, 4))
plt.plot(frequencies[:N // 2], np.abs(fft_values[:N // 2]))  # Only positive frequencies
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power")
plt.title("EEG Power Spectrum (Trial 1, Channel 2)")
plt.axvline(0.24, color="r", linestyle="--", label="Expected 0.24 Hz Peak")
plt.legend()
plt.show()
```
