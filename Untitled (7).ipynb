```python
import numpy as np
import os
from scipy.signal import butter, filtfilt
from mne.decoding import CSP
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split


MAX_EEG_CHANNELS = 8
MAX_SAMPLES = 32706  
# Sampling Frequency (Hz)
FS = 250  

# Recursive file search function
def find_npy_files(directory):
    npy_files = []
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith(".npy"):
                npy_files.append(os.path.join(root, file))
    return npy_files

# Function to standardize EEG data shape
def standardize_shape(data, max_channels=MAX_EEG_CHANNELS, max_samples=MAX_SAMPLES):
    num_channels, num_samples = data.shape
    
    # Adjust channels
    if num_channels < max_channels:
        pad_width = ((0, max_channels - num_channels), (0, 0))
        data = np.pad(data, pad_width, mode='constant')
    elif num_channels > max_channels:
        data = data[:max_channels, :]
    
    # Adjust samples
    if num_samples < max_samples:
        pad_width = ((0, 0), (0, max_samples - num_samples))
        data = np.pad(data, pad_width, mode='constant')
    elif num_samples > max_samples:
        data = data[:, :max_samples]

    return data

# Bandpass filter function for motor imagery (7-30 Hz)
def bandpass_filter(data, lowcut=7.0, highcut=30.0, fs=FS, order=5):
    nyq = 0.5 * fs  # Nyquist frequency
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, data)  # Apply forward and backward filter

# Load and preprocess EEG data
def load_and_preprocess(base_directory):
    data_list = []
    labels_list = []
    file_count = 0

    for trial_num in range(1, 7):
        trial_dir = os.path.join(base_directory, f"trial{trial_num}")
        print(f"Checking: {trial_dir}")

        # Search for .npy files recursively
        npy_files = find_npy_files(trial_dir)

        if not npy_files:
            print(f" No .npy files found in {trial_dir}")
            continue
        
        for file_path in npy_files:
            raw_data = np.load(file_path)

            if raw_data.ndim != 2:
                print(f" Skipping {file_path}: Expected 2D array but got {raw_data.shape}")
                continue

            print(f" Loaded {file_path} with shape {raw_data.shape}")

            # Standardize data shape
            processed_data = standardize_shape(raw_data)

            # Apply bandpass filtering along each channel
            filtered_data = np.apply_along_axis(bandpass_filter, axis=1, arr=processed_data)

            data_list.append(filtered_data)

            # Extract label from trial number
            label = trial_num - 1 
            labels_list.append(label)

            file_count += 1

    if file_count == 0:
        print(" No valid EEG files were processed!")
        return None, None

    # Convert to NumPy arrays
    X = np.array(data_list, dtype=np.float64)
    y = np.array(labels_list)

    print(f" Successfully processed {file_count} EEG files.")
    print(f" Final data shape: {X.shape}, Labels shape: {y.shape}")
    return X, y

# Run the function to load, standardize, and filter the EEG data
directory_path = 'breanna-data'
X, y = load_and_preprocess(directory_path)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
FS = 250  
TIME_WINDOW = 5  
# Convert to sample points
SAMPLE_WINDOW = TIME_WINDOW * FS  

# Function to plot a portion of EEG data
def plot_eeg_segment(eeg_data, title="EEG Signal Segment", time_window=TIME_WINDOW):
    num_channels, num_samples = eeg_data.shape
    # Ensure we dont exceed available data
    sample_window = min(time_window * FS, num_samples)  
    
    # Convert samples to time (seconds)
    time_axis = np.arange(sample_window) / FS  
    
    plt.figure(figsize=(12, 6))
    
    # Plot each channel
    for ch in range(num_channels):
        plt.plot(time_axis, eeg_data[ch, :sample_window] + ch * 10, label=f"Channel {ch+1}")  # Offset for visualization

    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude (V)")
    plt.title(title)
    plt.legend()
    plt.show()

# Plot a portion of EEG data for the first trial
plot_eeg_segment(X[0], title="Segment of EEG Signal (First 5 Seconds)")
```

```python
import numpy as np

# Define epoch length (1 second = 250 samples)
EPOCH_LENGTH = 250  # Samples per epoch (FS = 250 Hz)

def epoch_data(eeg_data, labels, epoch_length=EPOCH_LENGTH):
    """
    Segment EEG data into fixed-length epochs.
    
    Parameters:
    eeg_data (numpy array): Shape (trials, channels, samples)
    labels (numpy array): Shape (trials,)
    epoch_length (int): Number of samples per epoch (default: 250)

    Returns:
    epoched_data (numpy array): Shape (total_epochs, channels, epoch_length)
    epoched_labels (numpy array): Labels for each epoch
    """
    epoched_data = []
    epoched_labels = []

    for trial_idx in range(eeg_data.shape[0]):
        trial = eeg_data[trial_idx]
        label = labels[trial_idx]

        # Compute number of epochs per trial
        num_epochs = trial.shape[1] // epoch_length

        for i in range(num_epochs):
            start = i * epoch_length
            end = start + epoch_length
            epoch = trial[:, start:end]  # Extract segment

            if epoch.shape[1] == epoch_length:
                epoched_data.append(epoch)
                epoched_labels.append(label)

    # Convert to NumPy arrays
    epoched_data = np.array(epoched_data)
    epoched_labels = np.array(epoched_labels)

    print(f"Epoching completed: {epoched_data.shape[0]} epochs created.")
    return epoched_data, epoched_labels

# Apply epoching to our EEG data
X_epoched, y_epoched = epoch_data(X, y)

# Print shape to verify
print(f"Epoched Data Shape: {X_epoched.shape}, Epoched Labels Shape: {y_epoched.shape}")
```

```python
def baseline_correction(eeg_data):
    """
    Apply baseline correction by subtracting the mean of each epoch.
    
    Parameters:
    eeg_data (numpy array): Shape (epochs, channels, samples)

    Returns:
    corrected_data (numpy array): Baseline-corrected EEG data
    """
    corrected_data = eeg_data - np.mean(eeg_data, axis=2, keepdims=True)  # Subtract mean per epoch
    print("Baseline correction applied.")
    return corrected_data

# Apply baseline correction
X_corrected = baseline_correction(X_epoched)

# Print shape to verify
print(f"Baseline Corrected Data Shape: {X_corrected.shape}")
```

```python
# Import necessary libraries
from mne.decoding import CSP
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split

# Define number of CSP components to extract
N_CSP_COMPONENTS = 8  # Typically 3 per class

# Split into train & test sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X_corrected, y_epoched, test_size=0.2, random_state=42)

# Initialize CSP (Only works for binary classification)
csp = CSP(n_components=N_CSP_COMPONENTS, log=True, cov_est='epoch')

# Apply CSP transformation
X_train_csp = csp.fit_transform(X_train, y_train)
X_test_csp = csp.transform(X_test)

# Print transformed shape
print(f"CSP Transformed Train Data: {X_train_csp.shape}")
print(f"CSP Transformed Test Data: {X_test_csp.shape}")
```

```python
# Train LDA classifier
lda = LinearDiscriminantAnalysis()
lda.fit(X_train_csp, y_train)

y_pred = lda.predict(X_test_csp)

accuracy = (y_pred == y_test).mean() * 100
print(f"Classification Accuracy: {accuracy:.2f}%")
```
